# Setting the Environmet

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from pandas.core.common import flatten

from gensim.models import Word2Vec
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
import gensim
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cols_to_load = ['sentimen','tweet']
df_tweets = pd.read_csv('./combined_tweets.csv', usecols=cols_to_load)
sentiments = df_tweets.sentimen
tweets = df_tweets.tweet
df_tweets
```

# Tokenization


```python
def tokenize_tweet(tweet):
    # Pastikan nltk punkt sudah diunduh
    nltk.download('punkt', quiet=True)
    
    # Gunakan fungsi apply untuk menerapkan nltk.word_tokenize ke setiap baris dalam series
    tokenized_series = tweet.apply(nltk.word_tokenize)
    
    return tokenized_series
```

```python
tokenized_tweets = tokenize_tweet(tweets)
tokenized_tweets.head(5)
```

# Vectorization

## TF IDF

```python
vectorizer = TfidfVectorizer()
tweets_countvector = vectorizer.fit_transform(tweets)
```

```python
vectorizer.vocabulary_
```

```python
tweets_countvector.shape
```

```python
print(tweets_countvector)
```

```python
# Ubah scipy.sparse.csr_matrix menjadi pandas DataFrame
tweets_countvector = pd.DataFrame(tweets_countvector.toarray())
```

# Label Encoder for Sentiments

```python
# buat label encoder
labelencoder = LabelEncoder()

encoded_sentiments = labelencoder.fit_transform(sentiments)
print('Encoded labels:', encoded_sentiments)
print('Original labels:', labelencoder.inverse_transform(encoded_sentiments))
```

```python
encoded_sentiments = pd.DataFrame(encoded_sentiments, columns=['sentimen'])
encoded_sentiments.head(5)
```

# Merge Sentimen and Tweet Cols

```python
df_vectorized_tweets = pd.concat([encoded_sentiments, tweets_countvector], axis=1)
```
