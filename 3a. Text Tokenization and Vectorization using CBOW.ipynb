# Setting the Environmet

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from pandas.core.common import flatten

from gensim.models import Word2Vec
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import gensim
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.preprocessing import LabelEncoder
```

```python
cols_to_load = ['sentimen','tweet']
df_tweets = pd.read_csv('./combined_tweets.csv', usecols=cols_to_load)
sentiments = df_tweets.sentimen
tweets = df_tweets.tweet
df_tweets
```

# Tokenization


```python
tokenized_tweets =  tweets.apply(gensim.utils.simple_preprocess)
tokenized_tweets.head(5)
```

# Vectorization Using Word2Vec CBOW

## Model 1

```python
model_cbow = Word2Vec(tokenized_tweets,min_count=3, workers=3, window=5, sg=0)
```

```python
model_cbow.build_vocab(tokenized_tweets, progress_per=1000)
model_cbow.train(tokenized_tweets, total_examples=model_cbow.corpus_count, epochs=model_cbow.epochs)
```

```python
model_cbow.corpus_count
```

```python
model_cbow.save('./model_cbow.model')
```

```python
print(model_cbow.wv.most_similar('prabowo'))
print('-'*120)
print(model_cbow.wv.most_similar('jokowi'))
```

```python
print(model_cbow.wv['prabowo'])
print('-'*120)
print(model_cbow.wv['jokowi'])
```

```python
model_cbow.wv.key_to_index
```

```python
vectorized_tweets = tokenized_tweets.apply(lambda x: [model_cbow.wv[tweet] for tweet in x if tweet in model_cbow.wv.key_to_index])
vectorized_tweets = vectorized_tweets.apply(lambda x: list(flatten(x)))
vectorized_tweets.head(5)
```

# Label Encoder for Sentiments

```python
# buat label encoder
labelencoder = LabelEncoder()

encoded_sentiments = labelencoder.fit_transform(sentiments)
print('Encoded labels:', encoded_sentiments)
print('Original labels:', labelencoder.inverse_transform(encoded_sentiments))
```

```python
encoded_sentiments = pd.DataFrame(encoded_sentiments, columns=['sentimen'])
encoded_sentiments.head(5)
```

# Merge Sentimen and Tweet Cols

```python
df_vectorized_tweets = pd.concat([encoded_sentiments, vectorized_tweets], axis=1)
```
